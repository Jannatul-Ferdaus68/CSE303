```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('DataSheet.csv')

def plot_target_balance_bar(column_name):
    print(f"Distribution of '{column_name}':")
    counts = df[column_name].value_counts()
    print(counts)
    

    plt.figure(figsize=(6, 4))
    sns.barplot(x=counts.index.astype(str), y=counts.values, palette='Set2')
    plt.title(f"Class Distribution in '{column_name}'", fontsize=12)
    plt.xlabel(column_name, fontsize=10)
    plt.ylabel('Count', fontsize=10)
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

plot_target_balance_bar('Purchase_Made')
```

```python
import pandas as pd

def analyze_columns(df):
    for column in df.columns:
        print(f"\nColumn: {column}")
        print("-" * (9 + len(column)))
        
        col_data = df[column].dropna()  
        unique_values = col_data.nunique()

        if pd.api.types.is_numeric_dtype(col_data) and unique_values > 20:
            
            print(f"Type: Continuous Numerical")
            print(f"Mean: {col_data.mean():.2f}")
            print(f"Median: {col_data.median():.2f}")
            print(f"Variance: {col_data.var():.2f}")
            print(f"Standard Deviation: {col_data.std():.2f}")
        
        else:
            
            print("Type: Categorical")
            freq = col_data.value_counts()
            percent = col_data.value_counts(normalize=True) * 100
            
            result = pd.DataFrame({'Frequency': freq, 'Percentage': percent.round(2)})
            print(result)

analyze_columns(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def show_pie_chart(df, column_name):
    # Count values, excluding NaN by default
    counts = df[column_name].value_counts()
    labels = counts.index.astype(str)

    # Pie Chart
    plt.figure(figsize=(5, 5))
    plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title(f"Pie Chart of '{column_name}'")
    plt.axis('equal')  # Keep the pie chart circular
    plt.tight_layout()
    plt.show()

show_pie_chart(df, 'Group')
show_pie_chart(df, 'Customer_Segment')
show_pie_chart(df, 'Purchase_Made')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def show_distribution(df, column_name):
    
    counts = df[column_name].value_counts()
    
    labels = counts.index.astype(str)

    # Bar Chart
    plt.figure(figsize=(7, 4))
    plt.bar(labels, counts.values, color='lightgreen')
    plt.title(f"Bar Chart of '{column_name}'")
    plt.xlabel(column_name)
    plt.ylabel("Count")
    for i, value in enumerate(counts.values):
        plt.text(i, value + 0.5, str(value), ha='center')
    plt.tight_layout()
    plt.show()

    # Pie Chart
    plt.figure(figsize=(5, 5))
    plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title(f"Pie Chart of '{column_name}'")
    plt.axis('equal')
    plt.tight_layout()
    plt.show()

# Example use
show_distribution(df, 'Purchase_Made')

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Encode all categorical columns using LabelEncoder
def encode_categoricals(df):
    df = df.copy()  # avoid modifying the original
    le = LabelEncoder()
    for col in df.select_dtypes(include='object'):
        df[col] = le.fit_transform(df[col].astype(str))  # convert to string for safety
    return df

# Plot correlation heatmap
def plot_heatmap(df):
    corr = df.corr()
    plt.figure(figsize=(10, 6))
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title("Correlation Heatmap")
    plt.tight_layout()
    plt.show()

# Apply the functions
df_encoded = encode_categoricals(df)
plot_heatmap(df_encoded)

```
